# The Market for Lemons

**Inspired by:** Akerlof, G. A. (1978). The market for “lemons”: Quality uncertainty and the market mechanism. In _Uncertainty in economics_ (pp. 235-251). Academic Press.

A car has some intrinsic value $v$. The seller knows $v$, but the buyer does not. The buyer is willing to pay up to $1.5v$ for the car. How much should the buyer bid?

```python
from memo import memo
import jax.numpy as np
from matplotlib import pyplot as plt
from jax.scipy.stats.bernoulli import pmf as ber

from typing import no_type_check
```

```python
Value = np.linspace(0, 1, 101)

@no_type_check
@memo
def buyer():
    buyer: wants(payoff=
        (seller.quality * 1.5 - bid)  # 1.5x true value
        if seller.sell else 0         # only if transaction happens
    )
    buyer: chooses(bid in Value, to_maximize=EU[payoff])
    buyer: thinks[
        # The seller knows the quality; the buyer has uniform prior.
        seller: given(quality in Value, wpp=1),
        # The seller knows the bid.
        seller: knows(bid),
        # The seller sells iff bid >= quality.
        seller: chooses(sell in Bool, wpp=ber(sell, bid >= quality))
    ]
    return E[buyer.bid]

print(f'The rational buyer bids: ${buyer()}')
```
